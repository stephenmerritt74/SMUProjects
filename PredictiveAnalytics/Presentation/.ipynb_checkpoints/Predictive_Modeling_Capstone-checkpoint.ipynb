## Cluster Analysis for Store Formats

A company, currently comprised of 85 grocery stores, is looking to expand operations and open 10 new stores. With the opening of these new stores, the company is looking to diversify store formats to reduce product surpluses and shortages. We will first use cluster analysis to determine the optimal number of new store formats based on existing store sales percentages of the following categories: *Bakery, Deli, Frozen Food, Meat, Dry Grocery, General Merchandise, Dairy, Produce,* and *Dairy*.

Based on the aforementioned Cluster analysis using K-Means, the optimal number of new store formats is 3. We see below in figure 1-1 that the Median value of both the Adjusted Rand and Calinski-Harabasz indices are highest at 3.

## Classification Modeling for Store Format Assignment

Now that we have determined the optimal number of store formats, we’ll look to assign these formats to the 10 new stores. We’ll look at three different modeling techniques to assign formats to the new stores comparing demographic data of our existing stores to the demographic data of the new stores. Since this is a non-binary classification problem, the modeling techniques we’ll compare are a Decision Tree, Forest Model, and Boosted Model. While the accuracy scores of each model are very close, the **Boosted Model** has a far superior F1 statistic, so the Boosted Model is chosen. Please see Table 2-1 for performance comparison of the three models.

## Time Series Modeling for Sales Forecasts

Finally, we will attempt to predict the produce sales for 2016 for both our 85 existing and 10 new stores. We’ll use time series analysis to make these predictions. As can be seen in Figure 3-1, the time series for produce sales is Seasonal, with no discernible trend, with an error term. We will use an ETS (M, N, M) non-dampened model, along with a Seasonal ARIMA(0,1,1)(0,1,1)[12] using 6 months of holdout data to compare forecasts and fit statistics.

```python
Image(filename='/Users/Steve/SMUProjects/PredictiveAnalytics/Graphics/Fig 3-1.PNG', width = 500, height = 500)
```

```python
Image(filename='/Users/Steve/SMUProjects/PredictiveAnalytics/Graphics/Fig 3-1_2.PNG', width = 700, height = 700)
```

```python
Image(filename='/Users/Steve/SMUProjects/PredictiveAnalytics/Graphics/Fig 3-1_1.PNG', width = 500, height = 500)
```

After comparing the in-sample error measures for the different models, the fit statistics, and autocorrelation plots, the ETS(MNM) non-dampened Models is chosen to be most appropriate. The seasonal ARIMA(0, 1, 1)(0,1,1)[6] is close, but the plots in figure 3-2 above demonstrate that the ETS Model more closely models the 6-month holdout data. Also the RMSE for the ETS Model is lower at 760267.4 vs. 792197.3.

Utilizing the ETS(MNM) model, the final 2016 forecast numbers for both new and existing store produce sales can be found in Table 3-3 below.
